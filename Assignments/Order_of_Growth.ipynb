```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Setting valus for x axis
x = np.logspace(0,15,100,dtype=np.float128)
```

```python
# Function: 1
y_1 = np.ones(len(x))

# Function: n
y_n = np.array([i for i in x])

# Function: n^2
y_n_sq = np.square(x)

# Funtion: sqrt(n)
y_n_sqrt = np.sqrt(x)

# Function: n log(n)
y_nlogn = np.array([i * np.log(i) for i in x]) 

# Function: n^3
y_n_cube = np.array([i*i*i for i in x])

# Function: 2^n
y_2n = np.array([2**i for i in x])
```

```python
fig = plt.figure(figsize=(8, 6))
p = fig.add_subplot(1,1,1)
p.set_title('Order of growth Plot')
p.set_yscale('log')
p.set_xscale('log')
p.set_xlim(1,10**15)  
p.set_ylim(1,10**19)
p.plot(x,y_1,label='1')
p.plot(x,y_n,label='n')
p.plot(x,y_n_sq,label='n^2')
p.plot(x,y_n_sqrt,label='sqrt(n)')
p.plot(x,y_nlogn,label='nlogn')
p.plot(x,y_n_cube,label='n^3')
p.plot(x,y_2n,label='2^n')
plt.xlabel("Input Size")
plt.ylabel("Time Complexity")
p.legend()
```
